```python
# 예제3.7 - 상대 url을 절대 url로 변경하기
import requests
import lxml.html

response = requests.get('http://www.hanbit.co.kr/store/books/new_book_list.html')
response
```

```python
response.url
```

```python
root = lxml.html.fromstring(response.content)

# 모든 링크를 절대 URL로 변환합니다.
root.make_links_absolute(response.url)

# 선택자를 추가해서 명확한 선택을 할 수 있게 합니다.
# 책에는 제너레이터를 활용했지만, 직관적으로 이해하기 위해서는 list&append를 활용한 방법을 권장합니다.
url_list = []
for a in root.cssselect('.view_box .book_tit a'):
    url = a.get('href')
    url_list.append(url)
    
print(len(url_list))
print(url_list)
# 20개의 url에 대한 책 정보를 크롤링해보자
```

## 함수로 리팩터링

```python
# 예제3.8 - 웹 페이지에서 url 리스트 추출하기
def scrape_list_page():
    response = requests.get('http://www.hanbit.co.kr/store/books/new_book_list.html')
    root = lxml.html.fromstring(response.content)
    root.make_links_absolute(response.url)             # 절대 경로로 변경
    
    # 크롤링한 url 결과들을 리스트에 받고 반환하기
    url_list = []
    for a in root.cssselect('.view_box .book_tit a'):
        url = a.get('href')
        url_list.append(url)

    return url_list
```

```python
a = scrape_list_page()
print(len(a))
print(a)
```

# ch3.7.3 전체 url에 대해 크롤링하기

```python
# 예제 3.11 코드 변형
```

```python
# 전체 url에 대해 반복문으로 컨텐츠 수집
dict = {}

for i in tqdm_notebook(range(0, len(url_list))):
    # 전체 url에 대한 데이터 스크레이핑
    response = requests.get(url_list[i])
    root = lxml.html.fromstring(response.content)

    # 상세 페이지의 Response에서 책 정보를 dict로 추출합니다.
#     print('url:', response.url, '\n')
#     print('title:', root.cssselect('.store_product_info_box h3')[0].text_content(), '\n')
#     print('price:', root.cssselect('.pbr strong')[0].text_content(), '\n')
#     print('content:', '\n', [p.text_content() for p in root.cssselect('#tabs_3 .hanbit_edit_view p')], '\n')

    # 첫번째 url 컨텐츠를 딕셔너리에 저장
    ebook = {
            'url': response.url,
            'title': root.cssselect('.store_product_info_box h3')[0].text_content(),
            'price': root.cssselect('.pbr strong')[0].text_content(),
            'content': [p.text_content() for p in root.cssselect('#tabs_3 .hanbit_edit_view p')]
            }
    dict[i] = ebook
```
