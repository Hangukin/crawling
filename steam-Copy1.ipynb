```python
import sys
import os

import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains as AC

import chromedriver_autoinstaller

import re

import time
from time import sleep
from tqdm import tqdm
from tqdm import tqdm_notebook

import warnings
warnings.filterwarnings('ignore')
```

```python
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path)

driver.get("https://store.steampowered.com/")
time.sleep(3)
```

```python
element = driver.find_element_by_css_selector('#noteworthy_tab .pulldown_desktop')
element.click()
time.sleep(1)
element = driver.find_element_by_css_selector('#noteworthy_flyout .popup_menu_item')
element.click()
time.sleep(1)
```

```python
def page_scroll(driver):
    driver.execute_script("window.scrollTo(0, 19431049)")
    time.sleep(1)
    
n = 1
i = 0
while i < n: 
    page_scroll(driver) 
    i = i+1
```

```python
t_list = []
d_list = []
r_list = []
p_list = []

for i in tqdm_notebook(range(0,n+1)):
    #제목
    titles = driver.find_elements_by_css_selector('.title')
    temp_list=[]
    for title in titles:
        temp_list.append(title.text)
    t_list = t_list + temp_list 
    
    #날짜
    date_list = []
    dates = driver.find_elements_by_css_selector('.col.search_released.responsive_secondrow')
    for d in dates:
        release = d.text
        date_list.append(release)
    d_list = d_list + date_list
    
    #리뷰
    review_list = []
    for r in range(0,len(date_list)):
        reviews = driver.find_elements_by_xpath(f'//*[@id="search_resultsRows"]/a[{r+1}]/div[2]/div[3]/span')
        for items in reviews:
            review = items.get_attribute('data-tooltip-html')
        review_list.append(review)
    r_list = r_list + review_list
    
    #가격
    prices = driver.find_elements_by_css_selector('.col.search_price.responsive_secondrow')    
    price_list=[]
    for price in prices:
        won = price.text
        price_list.append(won)
    p_list = p_list + price_list
```

```python
df = pd.DataFrame(list(zip(t_list,d_list,r_list,p_list)), columns = ['제목','날짜','평가','가격'])
df
```

```python
df = df.drop_duplicates('평가')
df
```

```python
df.to_csv("steam_crawl.csv", encoding ='utf-8-sig')
```
